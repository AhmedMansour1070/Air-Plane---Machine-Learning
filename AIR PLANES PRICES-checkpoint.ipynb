```python
duration = list(data.Duration)

for i in range(len(duration)):
    if len(duration[i].split()) != 2:
        if 'h' in duration[i]:
            duration[i] = duration[i] + ' 0m'
        else:
            duration[i] = '0h ' + duration[i]
            
duration
duration_hour=[]
duration_minute=[]

for i in range(len(duration)):
    duration_hour.append(int(duration[i].split(sep='h')[0] ))
    duration_minute.append(int( duration[i].split('m')[0].split()[-1] ))

data['Duration_Hour'] = duration_hour
data['Duration_Minute'] = duration_minute
```

```python
durationminute = []
enc_copy = encodded_data.copy()
Date_of_Journey = data['Date_of_Journey']
Duration = data['Duration']
Total_Stops = data['Total_Stops']

enc_copy['Duration_Hour'] = duration_hour
enc_copy['Duration_Minute'] = duration_minute
durationminute[:] = [x / 100 for x in duration_minute]
new_duration = np.add(duration_hour, durationminute)
enc_copy['New_Duration'] = new_duration

data['Dep_hour'] = pd.to_datetime(data.Dep_Time ).dt.hour
data['Dep_Minute'] = pd.to_datetime(data.Dep_Time ).dt.minute

enc_copy['Dep_hour'] = data['Dep_hour']
enc_copy['Dep_Minute'] = data['Dep_Minute']

data['Journey_Day'] = pd.to_datetime(data.Date_of_Journey , format = '%d/%m/%Y').dt.day
enc_copy['Journey_Day'] = data['Journey_Day']


enc_copy= enc_copy.drop(['Duration_Hour' ,  'Duration_Minute' , 'Price'] , axis = 1)
train_x = enc_copy.values
train_y = data['Price']
train_x.shape
enc_copy
```

```python
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split( train_x, train_y, test_size=0.2, random_state=37)
```

```python
model = Sequential()
BatchNormalization()
model.add(Dense(12, activation = 'relu' , input_dim =27 ))
Dropout(0.25)
model.add(Dense(5 , activation = 'relu'))
Dropout(0.25)
model.add(Dense(3 , activation = 'relu'))

model.add(Dense(1))
model.summary()
train_x.shape
```

```python
from keras.callbacks import EarlyStopping

early_stop = EarlyStopping(monitor='val_loss' , patience= 3)
```

```python
model.compile(optimizer='adam' , loss = 'mse' , metrics=['Recall'])

model.fit(train_x , train_y, validation_split=0.10, epochs=100 , callbacks = [early_stop])
```

```python
pd.DataFrame(model.history.history).plot()
```

```python
plt.plot(model.history.history['recall'],label = 'loss')
plt.plot(model.history.history['val_loss'], label = 'val_recall')
plt.title('model Loss')
plt.ylabel('val_recall')
plt.xlabel('epoch')
plt.legend()
```

```python
result = model.evaluate(test_x, test_y)
result
```
